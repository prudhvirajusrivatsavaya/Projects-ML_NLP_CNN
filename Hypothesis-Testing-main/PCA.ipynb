```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
```

```python
# Perform PCA to extract principal components
pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
# Calculate the explained variance ratio for each component
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
explained_variance_ratio
```

```python
# Set a significance level (alpha)
alpha = 0.05
```

```python
# Perform hypothesis testing to determine the number of significant components
num_components = len(explained_variance_ratio)
significant_components = []

for i in range(num_components):
    # Calculate cumulative explained variance ratio up to component i
    cumulative_variance_ratio = np.sum(explained_variance_ratio[:i + 1])
    
    # Perform a hypothesis test to check if the explained variance is significant
    p_value = 1 - cumulative_variance_ratio
    if p_value < alpha:
        significant_components.append(i)
```

```python
# Plot the cumulative explained variance
plt.plot(np.cumsum(explained_variance_ratio))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance vs. Number of Components")
plt.grid()
plt.show()

print("Significant Components:", significant_components)
```
